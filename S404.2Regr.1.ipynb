### Warm up questions


Three design principles of Design of Experiments:


* replication: apply the same treatment to several experiment units.

> It reduces the variance of the sample mean, not variance of each observation.
    
    
* randomization: randomize the treatment assigned to experiment units and run experiments in random order if applicable.


>It helps to prevent lurk factors from distorting the comparison of treatments.
    
   
* blocking: divide experiment units into homogeneous/alike groups. Compare treatments in the same groups.


> Isolate the influence of known factors from the comparison of treatments.

### What is regression?

Who coined the term regression?

**Francis Galton**


* The term "regression" was coined by Francis Galton in the nineteenth century to describe a biological/genetic phenomenon. 


* The phenomenon was: the heights of descendants of tall ancestors tend to regress down towards a normal average.


* This phenomenon is also known as regression toward the mean.


### Statistical definition of regression function


* Let $Y$ be the height of the first son (if any), and $X$ be the height of the father of a randomly sampled pair from a biological population.
                  
                  
* Due to inheritance/genetics plus randomness, the tall fathers give birth of tall sons on average but not in every case.


* The average height of the son given the height of his father is

> $ g(x) = E(Y| X = x)$.


* Mathematical, it is a function of $x$. Statistically, we call $g(x)$ "regression function".


### Data and fitting of the regression function.


* When $X$ and $Y$ are highly relevant, we may collect a large amount of data (observational or experimental) ($x_i, y_i$) based on which we hope to get a good idea on regression function $g(x)$.


* This is a typical statistical project/task.


* The task is much simpler if a reasonable structure is imposed on how $x$ and $y$ are connected.


* This leads to the model assumptions unopposed in general.

### Linear regression model.

* In many applications, a response variable $Y$ is seen related to a design-variable/covariate/prediction-variable/independent-variable/whatever-variable.


* it is further reasonable to assume that 

> $ Y = \alpha + \beta X + \epsilon$

> for some unknown and nonrandom constants $\alpha$ and $\beta$.


* In DoE, $X$ may not be random but carefully chosen. In observational studies, 
$X$ is usually random.


* We assume $X$ and $\epsilon$ independent in this course.


* If not declared, we regard $X$ as not random. Theoretically, it amounts to say the analysis is "conditional on $X$".

### Assumption on the error term


* Particularly in introductory courese, we assume $\epsilon \sim N(0, \sigma^2)$.


> Theoretical conclusions you are taught are solid under this assumption.


* One may weaken it to $E(\epsilon) = 0$ and $Var(\epsilon) = \sigma^2 < \infty$. 


> conclusions you are taught remain a good approximation under these two assumptions in absence of normality.


* We also call $\epsilon$ experiment error and $\sigma^2$ error variance.

### The non-intuitive reason for being called "linear"


* Under the model assumption, 
$E(Y|X= x) = \alpha + \beta x$ with $\alpha$ and $\beta$ 
values unknown.


* We say it is a linear regression because $E(Y|X= x)$ is linear in $\alpha$ and $\beta$.


* Based on this definition,


> If $ Y = \alpha + \beta_1 X + \beta_2 X^2 + \epsilon$, the model is still linear. 
It is linear in $\alpha, \beta_1$ and $\beta_2$, but not in $x$.


>  $ Y = (\alpha + \beta X)^2 + \epsilon$ is not linear. 


> $ Y = \alpha + \beta^3 X + \epsilon$ is not linear, due to a silly choice of $\beta^3$.


**It matters little if you do not remember this definition unless you wish to be a pro.**

### Multiple regression

* More often than not, there are multiple prediction variables so that data must be described as


> $
y_i = \beta_0 + \beta_1 x_{i1} + \cdots + \beta_k x_{ik} + \epsilon_i, ~~i=1,2, \ldots, N
$

> with independent observations on $N$ units and $k$ prediction variables. 



> e.g. The height of a person is widely known related to heights of father and mother, gender, nutrition and more. Multiple regression model is most sensible.



* These $N$ equations formed by $N$ units can be written in matrix form
$
 {\bf y} = {\bf X} {\mathbf \beta} + {\mathbf \epsilon}.
$

>  ${\bf y}$ and ${\mathbf \epsilon}$ are vectors of length $N$,

>  ${\bf X}$ a design matrix of size $N \times (k+1)$ and 

>  ${\mathbf \beta}$ is the regression coefficient vector of length $k+1$.



### Multiple regression model

### Estimation of $\beta$


* Under a linear regression model, we may estimatie ${\mathbf \beta}$ by
$\hat {\mathbf \beta}$ that minimizes
$$
 ({\bf y} - {\bf X} {\mathbf \beta})^T ({\bf y} - {\bf X} {\mathbf \beta}) 
 = \sum_{i=1}^N (y_i - {\bf x_i} {\mathbf \beta})^2.
$$


* The optimization solution is given by
$$
 \hat {\mathbf \beta} = ( {\bf X}^T  {\bf X})^{-1}  {\bf X}^T {\bf y}.
$$



**We call this estimate the least-squares estimate (estimator).**


### Some terminologies under linear model

* We call $ {\bf r} = {\bf y} - {\bf X} \hat {\mathbf \beta}~$ **residuals**;


* We call
$\hat {\bf y} = {\bf X} \hat {\mathbf \beta}~$ **fitted values**;


* We call $\hat {\mathbf \beta}~$ the **least-squares estimate**.
 

> Under normality assumption on experiment error $\epsilon$, 
$\hat {\mathbf \beta}~$ is also the maximum likelihood
estimate of ${\mathbf \beta}$.

### An extra dose of mathematical statistics


* estimator of a parameter: it is a **function** of data


* estimate: the value of the estimator once evaluated given data.


* estimation: the statistical notion of estimating parameters in a statistical model.

### Algebraic facts in the context of linear model


* The least squares estimator is given by
$$
 \hat {\mathbf \beta} = ( {\bf X}^T  {\bf X})^{-1}  {\bf X}^T {\bf y}.
$$



* Given this knowledge, we find
\begin{align*}
{\bf X}^T ( {\bf y} - {\bf X} \hat {\mathbf \beta} )
& = {\bf X}^T {\bf y} - {\bf X}^T {\bf X} ( {\bf X}^T {\bf X})^{-1}  {\bf X}^T {\bf y}\\
& = {\bf X}^T {\bf y} - {\bf X}^T {\bf y}\\
&= 0.
\end{align*}


* Using ${\bf r} = {\bf y} - {\bf X} \hat {\mathbf \beta}$, the above conclusion can be written as ${\bf X}^T {\bf r} = 0$.


> Nothing in the residual is (linearly) related to ${\bf X}$.


### Algebraic facts in the context of linear model

* Denote
$\bar {\bf x} = N^{-1} \sum_{i=1}^N  {\bf x}_i $
and
$\bar {y} = N^{-1} \sum_{i=1}^N  y_i $.


* Let
$$\bar {\bf y} = 
\left (
\begin{matrix}
\bar y\\ \bar y\\ \vdots \\ \bar y
\end{matrix}
\right ),
\hspace{1em}
\bar {\bf X} = 
\left (
\begin{matrix}
\bar {\bf x}\\ \bar {\bf x}\\
\vdots \\ \bar {\bf x}
\end{matrix}
\right )
\hspace{1em}
\text{  and } 
\hspace{1em}
{\bf X} - \bar {\bf X}
=
\left (
\begin{matrix}
{\bf x}_1 - \bar {\bf x}\\ 
{\bf x}_2 - \bar {\bf x}\\
\vdots \\ 
{\bf x}_n - \bar {\bf x}
\end{matrix}
\right ).
$$


* Recall ${\bf X}^T ( {\bf y} - {\bf X} \hat {\mathbf \beta} ) = 0$.


* Becasue the first row of ${\bf X}^T$ is $(1, 1, \ldots, 1)$, the above equation contains the result:
$$ \bar y - \bar{\bf x} \hat {\mathbf \beta} =0.$$


* the predicted response value at the average covariate value ${\bf x}$ is the sample mean $\bar y$.


### Algebraic facts in the context of linear model


* We may write
\begin{align*}
{\bf y} - \bar{\bf y}
& = ({\bf y} - \bf X \hat {\mathbf \beta})
   + (\bf X \hat {\mathbf \beta} - \bar{\bf y})\\
& = ({\bf y} - \bf X \hat {\mathbf \beta})
+ (\bf X - \bar{\bf X}) \hat {\mathbf \beta}.
\end{align*}



* With this decomposition, we find
\begin{align*}
({\bf y} - \bar{\bf y})^T ({\bf y} - \bar{\bf y})
& = 
({\bf y} - \bf X \hat {\mathbf \beta})^T({\bf y} - \bf X \hat {\mathbf \beta})
 + 
\hat {\mathbf \beta}^T (\bf X - \bar{\bf X})^T (\bf X - \bar{\bf X}) \hat {\mathbf \beta}\\
& = 
({\bf y} - \hat{\bf y})^T({\bf y} - \hat{\bf y})
 + 
(\hat{\bf y} - \bar{\bf y} )^T(\hat{\bf y} - \bar{\bf y} ).
\end{align*}


* The cross-term on the RHS equals 0. To be given as an assignment problem.


**A student may simply give above proof without knowing why the cross-term equals 0. It is hard for TA or myself to know whether s/he actually has the knowledge.**

### Interpretting the decomposition:


* Watch closely on the decomposition:
\begin{align*}
({\bf y} - \bar{\bf y})^T ({\bf y} - \bar{\bf y})
& =  
\hat {\mathbf \beta}^T (\bf X - \bar{\bf X})^T (\bf X - \bar{\bf X}) \hat {\mathbf \beta}
+
({\bf y} - \bf X \hat {\mathbf \beta})^T({\bf y} - \bf X \hat {\mathbf \beta})
\end{align*}
or it is
\begin{align*}
({\bf y} - \bar{\bf y})^T ({\bf y} - \bar{\bf y})
& = (\hat {\bf y} - \bar {\bf y})^T (\hat {\bf y} - \bar {\bf y})
+ {\bf r}^T {\bf r}
\end{align*}


* the LHS is the total variation in ${\bf y}$.


* the first term the RHS the variation in $\hat {\bf y}$.


* the second term is the variation in residual ${\bf r}$.



* This decomposition results in the analysis of variance table (ANOVA):

### ANOVA Table for General Linear Model


* ANOVA Table
\begin{matrix}
\mbox{Source} & \mbox{Df} & \mbox{SS} & \mbox{MSS} & \mbox{F} \\
\mbox{Regr} & k & 
(\hat {\bf y} - \bar {\bf y})^T (\hat {\bf y} - \bar {\bf y}) 
&  (\hat {\bf y} - \bar {\bf y})^T (\hat {\bf y} - \bar {\bf y}) /k \\
\mbox{Residual/error} & N-k-1 & 
({\bf y}-{\bf X}\hat{\bf \beta})^T({\bf y}-{\bf X}\hat{\bf \beta}) &
({\bf y}-{\bf X}\hat{\bf \beta})^T({\bf y}-{\bf X}\hat{\bf \beta})/(N-k-1) & \\
\mbox{Total} & N-1 & {\bf y}^T{\bf y}- \bar{\bf y}^T\bar{\bf y}
\end{matrix}


* The top-right entry of F should be
$$
F = 
\frac{(\hat {\bf y} - \bar {\bf y})^T (\hat {\bf y} - \bar {\bf y}) /k }
{({\bf y}-{\bf X}\hat{\bf \beta})^T({\bf y}-{\bf X}\hat{\bf \beta})/(N-k-1)}
=\dfrac{\mbox{MSS}_{regr}}{\mbox{MSS}_{error}}.
$$


### How helpful is the regresson function? 


* Let
$$
R^2 
= \frac{ 
(\hat{\bf y}-\bar{\bf y})^T(\hat{\bf y}-\bar{\bf y})}
{({\bf y} - \bar {\bf y})^T ({\bf y} - \bar {\bf y}) }.
$$


* The closer is the fitted value and the observed value ($\hat {\bf y}$ and ${\bf y}$),
the larger is $R^2$.


* This value is the percentage of variations in ${\bf y}$ 
explained by the linear regression ($\hat {\bf y}$) according to
ANOVA table.


* A higher $R^2$ indicates a better fit.

### Estimate $\sigma^2$: the error variance.


* Let us refresh the memory of the linear model:
$$
Y = \beta_0 + \beta_1 x_{1} + \cdots + \beta_k x_{k} + \epsilon.
$$
with $\sigma^2 = var(\epsilon)$.


* We estimate $\sigma^2$ by 
$$
\hat \sigma^2 
= 
\frac{ 
{\bf r}^T {\bf r}}{N-k-1}
=
\frac{
({\bf y}-\hat{\bf y})^T({\bf y}-\hat{\bf y})}{(N-k-1)}.
$$


* Rational: under model assumptions, we have
$var(\hat \sigma^2) = \sigma^2$.

### Confidence interval for unknown parameters


* Under linear model, we have methods for estimating unknown parameters
${\bf \beta}$ and $\sigma^2$.



* Likely, the true values of ${\bf \beta}$ and $\sigma^2$ 
do not equal to but close to $\hat{\bf \beta}$ and $\hat{\sigma}^2$.



* Confidendence interval (region) is to draw a regions
around $\hat \theta$, such that it has a good chance to include
$\theta$.

> $\theta$ here is generic: ${\bf \beta}$, $\sigma^2$, ${\bf \beta} + 2\sigma$, ${\bf x} {\mathbf \beta}$ for instances.



* A general form of CI with confidence $(1-\alpha)$ of parameter $\theta$ is
$$ \hat \theta \pm c_{1-\alpha/2} \sqrt{\widehat{var}(\hat \theta)}.$$


**Important: The targets of CI or CR are unknown parameters.**

### Confidence interval for ${\bf x} {\mathbf \beta}$



* Consider $\theta = {\bf x} {\mathbf \beta}$, the expected response
value $y$ whose predictor equals ${\bf x}$.

> for instance, the expected height of a son whose father is of average height.


* We estimate it by $\hat \theta = \hat y = {\bf x} \hat{\mathbf \beta}$,
namely fitted value.


* A 95% CI for $\theta = {\bf x} {\mathbf \beta}$ is 

$$
{\bf x} \hat{\mathbf \beta} \pm 
t_{n-k-1, 0.975} \sqrt{\widehat{var}({\bf x} \hat{\mathbf \beta})}.
$$

> here $c_{1-\alpha/2}$ becomes $t_{n-k-1, 0.975}$, the 97.5% 
quantile of the t-distribution with $n-k-1$ degrees of freedom;


> because $var({\bf x} \hat{\mathbf \beta}) = {\bf x}({\bf X}^T{\bf X})^{-1}{\bf x}^T \sigma^2$, we let
$
\widehat{var}({\bf x} \hat{\mathbf \beta})
={\bf x}({\bf X}^T{\bf X})^{-1}{\bf x}^T \hat{\sigma}^2.
$


**Within ${\bf x} {\mathbf \beta}$, ${\bf x}$ is known, ${\mathbf \beta}$ is the unknown parameter.**

### A test for the usefulness of the linear model.



* If $H_0$ holds (the linear model is useless, or preditors are useless), we would find

$$ \mbox{MSS}_{regr} \approx \mbox{MSS}_{error}$$.



* Based on this logic, we introduced

$$
F = 
\frac{(\hat {\bf y} - \bar {\bf y})^T (\hat {\bf y} - \bar {\bf y}) /k }
{({\bf y}-{\bf X}\hat{\bf \beta})^T({\bf y}-{\bf X}\hat{\bf \beta})/(N-k-1)}
=\dfrac{\mbox{MSS}_{regr}}{\mbox{MSS}_{error}}.
$$


* Most of you know that this is called F-statistic.

## F-test for the usefulness of the model


* We first compute the value of $F$, and denote it as $F_{obs}$.


* If $F_{obs}$ is large, it sheds doubt in $H_0: \beta_1 = \beta_2 = \cdots = \beta_k = 0$.


* Yet we need a **reference distribution** to judge the largeness of $F_{obs}$,


* If $H_0$ holds, the distribution of $F$ statistic is F-distribution with $k$ and $N-k-1$ degrees of freedom. 


* If $H_0$ is violated, then the size of $F$ shifts toward the upper end of the F-distribution. 



* The chance of observing stronger evidence against $H_0$, when $H_0$ is the true, is given by
$
P\{ F > F_{obs}\}
$.


* The probability calculation depends on $F_{obs}$, $N$ and $k$ and is free from $(x_i, y_i)$ once $F_{obs}$ is given.

```r
## How large F-random variables get when k=3, N-k-1 = 100?

ff = rf(5000, 3, 100)
print(c(mean(ff), var(ff)))
options(repr.plot.width=10, repr.plot.height=5)
hist(ff, 30)
```

### Judgment call for the F-test.


* Suppose we indeed have $k=3$ and $N-k-1=100$.


* Out of 5000 experiments, if $H_0$ is true, none of the $F$ values exceed $7$.


* Based on this experience, I will be skeptical if someone says the model is useless
(i.e. $H_0$ is true) if $F_{obs} = 8$.


* Mathematically, we make this decision procedure more formal via the notion of p-value.

### P-value of a test


* We call $P\{ F > F_{obs}| H_0\}$ the **p-value** of **the test** against $H_0$.


* A somewhat artificial standard such as $0.05$ is commonly adopted as the cut-off point. 


* When a p-value is lower than $0.05$, we reject the null hypothesis at level $0.05$. 


**The use of p-value is widely criticized recently**.


* The concept of p-value is not well defined. In our example, it is meaningful only after we have accepted the linear model assumption and the use of F-test. These leave a big room for a scientist to manipulate their analysis.

### Should we use F-test or t-test?


* The answer depends on the question. 


* The discussion in the context of "is the model useful"


* F-test examines whether the whole model is useful or not.


* t-test focuses on if a specific predictor is useful or not?


* When there is only one predictor in an application, these two are the same. 


* You may use either one.


* Other than potential round-off error, the conclusion of two tests should be the same.


**In many statistical problems, there exist many valid and no-equivalent approaches. Their conclusions are not necessarily identical.**

```r
### Illustration with the current data example.

xx = read.table("JeffData1a14.txt")
x = as.matrix(xx)
y = x[, 5]
x = cbind(rep(1, 51), x[,1:4])
betahat = solve(t(x)%*%x)%*%(t(x)%*%y)


### The model contains beta[0] and beta[1]

### Let answer the question on the usefulness of beta[1] 
##          which it is the second entrance

beta.var = (solve(t(x)%*%x)*sigmahat^2)[2,2]

tt = abs(betahat[2]/(beta.var)^.5)

pvalue = 2 * pt(tt, 51-4-1, lower.tail = F)
### two-sided alternative

print(c(tt, pvalue))

```
